```python
x=data.drop(280,axis=1)
y=data[[280]]
```

```python
X_train_full,X_test,y_train_full,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
X_train,X_valid,y_train,y_valid=train_test_split(X_train_full,y_train_full,test_size=0.2,random_state=1)
```

```python
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) #to finding transformation variables and transforming training data
X_val_scaled = scaler.transform(X_valid)  #using those values transforming validation dataset
X_test_scaled = scaler.transform(X_test)
```

```python
X_train.shape[1:]
```

```python
def model_building(n_hidden=7,n_neurons=280,learning_rate=3e-3):
    model=keras.models.Sequential()
    model.add(keras.layers.Dense(n_neurons,activation="relu",input_shape=(280,)))
    for i in range(n_hidden):
        n_neurons=int(n_neurons*0.66)
        model.add(keras.layers.Dense(n_neurons,activation="relu"))
    model.add(keras.layers.Dense(1))
    optimizer=keras.optimizers.SGD(learning_rate)
    model.compile(loss="mean_squared_logarithmic_error",optimizer=optimizer)
    return model
```

```python
built_model=keras.wrappers.scikit_learn.KerasRegressor(model_building)
```

```python
#built_model.summary()
```

```python
built_model.fit(X_train_scaled,y_train,epochs=20,validation_data=(X_val_scaled,y_valid))
```

```python
#built_model.history
```

```python
mse_test = built_model.score(X_test_scaled, y_test)
```

```python
y_test_p=built_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_test_p))
```

```python
def model_building(n_hidden=7,n_neurons=280,learning_rate=3e-3):
    model=keras.models.Sequential()
    model.add(keras.layers.Dense(n_neurons,activation="relu",input_shape=(280,)))
    for i in range(n_hidden):
        n_neurons=int(n_neurons*0.66)
        model.add(keras.layers.Dense(n_neurons,activation="relu"),kernel_regulaizer=keras.regularizes.l2(0.01))
    model.add(keras.layers.Dense(1))
    optimizer=keras.optimizers.SGD(learning_rate)
    model.compile(loss="mean_squared_logarithmic_error",optimizer=optimizer)
    return model
```

```python
built_model_RE=keras.wrappers.scikit_learn.KerasRegressor(model_building)
built_model_RE.fit(X_train_scaled,y_train,epochs=20,validation_data=(X_val_scaled,y_valid))
```
